```python
import numpy as np
import pandas as pd
from collections import Counter
import textdistance 
import re
```

# Consider the word file 

```python
words = []
with open('text book.txt', 'r', encoding='utf-8') as f:
    data = f.read()
    data = data.lower()
    word = re.findall('\w+', data)
    words += word
```

# Find the frequency of words

```python
word_freq_dict = Counter(words)
```

```python
word_freq_dict
```

```python
word_freq_dict.most_common(10)                   #left = keys, right = values 
```

# Find relative frequency of words
Formula: Probability(word) = Frequency(word)/Total count of words

```python
word_freq_dict.values()
```

```python
Total_words_freq = sum(word_freq_dict.values())
```

```python
Total_words_freq
```

```python
probs = {}
for k in word_freq_dict.keys():
    probs[k] = word_freq_dict[k] / Total_words_freq
```

```python
probs             #Helped us find the probability of each word using the formula
```

# How to find the similar words

```python
def autocorrect(word):        #Hel
    word = word.lower()
    if word in probs:
        print('the word is already there', word)
    else:
        similarities = [1-(textdistance.Jaccard(qval=2)).distance(w,word) for w in word_freq_dict.keys()]
        df = pd.DataFrame.from_dict(probs,orient='index').reset_index()
        df = df.rename(columns={'index': 'Word', 0: 'Prob'})
        df['Similarity'] = similarities
        output = df.sort_values(['Similarity','Prob'],ascending=False).head(7)
        return(output)
        
autocorrect('mov')
```
